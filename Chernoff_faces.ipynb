```python
from ChernoffFace import *
import numpy
import matplotlib.cm
```

```python
# Генерируем данные
numpy.random.seed(32)
data = numpy.random.rand(16, 18)
```

```python
# Строим лица Чернова
fig = chernoff_face(data=data, 
                    titles=[str(x) for x in list(range(len(data)))], 
                    color_mapper=matplotlib.cm.Pastel1)

# Display
fig.tight_layout()
matplotlib.pyplot.show()
```

```python
# Рассмотрим еще один пример (датасет – данные об арестах в США)
dfData=load_usa_arrests_data_frame()
dfData.head()
```

```python
# Проведем масштабирование
dfData2 = variables_rescale(dfData)
dfData2.head()
```

```python
fig = chernoff_face(data=dfData2,
                    n_columns=5,
                    long_face=False,
                    color_mapper=matplotlib.cm.tab20c_r,
                    figsize=(12, 12), dpi=200)
fig.tight_layout()
matplotlib.pyplot.show()
```
