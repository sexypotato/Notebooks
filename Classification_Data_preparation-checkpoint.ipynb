```python
mapper = {'year': 365, 'month': 30, 'week': 7, 'day': 1}
```

```python
digit = data.AgeuponOutcome.str.split(' ').str[0].astype(float)
```

```python
calen = data.AgeuponOutcome.str.split(' ').str[1].str.rstrip('s').map(mapper)
```

```python
data['Age'] = digit * calen
```

```python
data.drop(['AgeuponOutcome'], axis=1, inplace=True)
```

###  

######    

```python
mapper
```

```python
data['Has_name'] = pd.isnull(data.Name) * 1
```

```python
data.Name = data.Name.fillna('Unknown')
```

```python
data['Name_len'] = data.Name.str.len()
```

```python
name_freq = data.Name[data.Name != 'Unknown'].value_counts(normalize=True).to_dict()
```

```python
data['NameFreq'] = data.Name.map(name_freq)
```

```python
data.drop(['Name'], axis=1, inplace=True)
```

###### 

```python
data.SexuponOutcome = data.SexuponOutcome.replace('Unknown', 'Unknown Unknown')
data['SexStatus'] = data.SexuponOutcome.str.split(' ').str[0]
data.SexStatus = data.SexStatus.replace({'Neutered': 'Flawed', 'Spayed': 'Flawed'})
```

```python
data['Sex'] = data.SexuponOutcome.str.split(' ').str[1]
```

```python
data.drop(['SexuponOutcome'], axis=1, inplace=True)
```

###### 

```python
data
```

```python
data['IsBiColor'] = data.Color.str.contains('/') * 1
```

```python
data.drop(['Color'], axis=1, inplace=True)
```

```python
data['MixBreed'] = data.Breed.str.endswith('Mix') * 1
data['Domestic'] = data.Breed.str.contains('Domestic') * 1
data['Shorthair'] = data.Breed.str.contains('Shorthair') * 1
```

```python
data.drop('Breed', axis=1, inplace=True)
```

```python
data
```

######   

```python
dates = data.DateTime.astype(np.datetime64)
```

```python
data['Year'] = dates.apply(lambda x: x.year)
data['Month'] = dates.apply(lambda x: x.month)
data['Day'] = dates.apply(lambda x: x.day)
data['Hour'] = dates.apply(lambda x: x.hour)
data['Weekday'] = dates.apply(lambda x: x.weekday())
```

```python
data.drop(['DateTime'], axis=1, inplace=True)
```

```python
data.dtypes
```

######     dummies

```python
data = pd.get_dummies(data, columns=['SexStatus', 'Sex'])
```

######   

```python
for col in data.columns:
    if set(data[col]) == set((0,1)):
        data[col] = data[col].astype(np.int8)
```

######   

```python
data.Age = data.Age.fillna(data.Age.mean())
```

```python
data.NameFreq = data.NameFreq.fillna(data.NameFreq.mean())
```

######     

```python
train = data[data.Test == 0].reset_index(drop=True)
```

```python
test = data[data.Test == 1].reset_index(drop=True)
```
