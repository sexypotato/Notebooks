```python
import matplotlib
import matplotlib.pyplot as plt
```

```python
# The plot() function is used to plot a line graph. 
# Для построения линейного графика используется функция plot()
x = [1, 5, 10, 15, 20]
y1 = [1, 7, 3, 5, 11]
y2 = [4, 3, 1, 8, 12]
plt.figure(figsize=(12, 7))
plt.plot(x, y1, 'o-r', alpha=0.7, label='first', lw=5, mec='b', mew=2, ms=10)
plt.plot(x, y2, 'v-.g', label='second', mec='r', lw=2, mew=2, ms=12)
plt.legend()
plt.grid(True)
```

```python
# The fill_between() function is used to fill areas. 
# Для заливки областей используется функция fill_between()
```

```python
import numpy as np
x = np.arange(0.0, 5, 0.01)
y = np.cos(x*np.pi)

# Display a chart with a fill
# Отобразим график с заливкой
plt.plot(x, y, c = 'r')
plt.fill_between(x, y)
```

```python
# Two-color fill option. / Вариант двухцветной заливки:
plt.plot(x, y, c = 'r')
plt.grid()
plt.fill_between(x, y, where=y>=0, color='g', alpha=0.3)
plt.fill_between(x, y, where=y<=0, color='r', alpha=0.3)
```

```python
# Stair Graph. / Ступенчатый график
x = np.arange(0, 7)
y = x
where_set = ['pre', 'post', 'mid']
fig, axs = plt.subplots(1, 3, figsize=(15, 4))
for i, ax in enumerate(axs):
 ax.step(x, y, 'g-o', where=where_set[i])
 ax.grid()
```

```python
# Stack chart. / Стековый график
# The stackplot() function is used to build a stack plot. 
# Для построения стекового графика используется функция stackplot().
# Its essence is that the graphs are displayed one above the other, and each next one is the 
# sum of the previous one and the given one.
# Суть его в том, что графики отображаются друг над другом, и каждый следующий является суммой 
# предыдущего и заданного

x = np.arange(0, 11, 1)
y1 = np.array([(-0.2)*i**2+2*i for i in x])
y2 = np.array([(-0.4)*i**2+4*i for i in x])
y3 = np.array([2*i for i in x])
labels = ['y1', 'y2', 'y3']
fig, ax = plt.subplots()
ax.stackplot(x, y1, y2, y3, labels=labels)
ax.legend(loc='upper left')
```

```python
# Stem-график
# Visually, the stem chart looks like a set of lines from the point c
# coordinates (x, y) to the baseline, at the top of which a marker is placed 
# Визуально stem-график выглядит как набор линий от точки с координатами (x, y) до базовой линии, 
# в верхней точке которой ставится маркер

x = np.arange(0, 10.5, 0.5)
y = np.array([(-0.2)*i**2+2*i for i in x])
plt.stem(x, y)
```

```python
# Scatter Plot. / Точечный график (Диаграмма рассеяния)
# The scatter() function is used to build a scatterplot. In its simplest form, it can be 
# obtained by passing the x and y coordinates to the scatter() function
# Для построения диаграммы рассеяния используется функция scatter(). В простейшем виде её можно 
# получить, передав функции scatter() координаты x и y

x = np.arange(0, 10.5, 0.5)
y = np.cos(x)
plt.scatter(x, y)
```

```python
# An example demonstrating working with color and size
# Пример, демонстрирующий работу с цветом и размером

import matplotlib.colors as mcolors
bc = mcolors.BASE_COLORS
x = np.arange(0, 10.5, 0.25)
y = np.cos(x)
num_set = np.random.randint(1, len(mcolors.BASE_COLORS), len(x))
sizes = num_set * 35
colors = [list(bc.keys())[i] for i in num_set]
plt.scatter(x, y, s=sizes, alpha=0.4, c=colors, linewidths=2, 
edgecolors='face')
plt.plot(x, y, 'g--', alpha=0.4)
```

```python
# Bar charts. / Столбчатые диаграммы
# Bar charts are good for visualizing categorical data.
# Для визуализации категориальных данных хорошо подходят столбчатые диаграммы. 
# To build them, the following functions are used / Для их построения используются функции:
# • bar() — is a vertical bar chart /вертикальная столбчатая диаграмма;
# • barh() — is a horizontal bar chart /горизонтальная столбчатая диаграмма.

np.random.seed(123)
groups = [f'P{i}' for i in range(7)]
counts = np.random.randint(3, 10, len(groups))
plt.bar(groups, counts)
```

```python
# An example demonstrating how to work with bar() parameters
# Пример, демонстрирующий работу с параметрами bar()

import matplotlib.colors as mcolors
bc = mcolors.BASE_COLORS
np.random.seed(123)
groups = [f'P{i}' for i in range(7)]
counts = np.random.randint(0, len(bc), len(groups))
width = counts*0.1
colors = [['r', 'b', 'g'][int(np.random.randint(0, 3, 1))] for _ in 
counts]
plt.bar(groups, counts, width=width, alpha=0.6, bottom=2, color=colors, 
edgecolor='k', linewidth=2)
```

```python
# Group bar charts / Групповые столбчатые диаграммы
# Using prepared data in a certain way, you can build group diagrams
# Используя определенным образом подготовленные данные, можно строить групповые диаграммы

cat_par = [f'P{i}' for i in range(5)]
g1 = [10, 21, 34, 12, 27]
g2 = [17, 15, 25, 21, 26]
width = 0.3
x = np.arange(len(cat_par))
fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, g1, width, label='g1')
rects2 = ax.bar(x + width/2, g2, width, label='g2')
ax.set_title('Пример групповой диаграммы')
ax.set_xticks(x)
ax.set_xticklabels(cat_par)
ax.legend()
```

```python
# Chart with errorbar element /Диаграмма с errorbar-элементом
# Errorbar элемент позволяет задать величину ошибки для каждого элемента графика. / 
# The Errorbar element allows you to set the error value for each plot element.
# Для этого используются параметры xerr, yerr и ecolor, первые два определяют величину ошибки, 
# последний — цвет
# For this, the xerr, yerr and ecolor parameters are used, the first two define the error value, 
# the last one defines the color

np.random.seed(123)
rnd = np.random.randint
cat_par = [f'P{i}' for i in range(5)]
g1 = [10, 21, 34, 12, 27]
error = np.array([[rnd(2,7),rnd(2,7)] for _ in range(len(cat_par))]).T
fig, axs = plt.subplots(1, 2, figsize=(10, 5))
axs[0].bar(cat_par, g1, yerr=5, ecolor='r', alpha=0.5, edgecolor='b', linewidth=2)
axs[1].bar(cat_par, g1, yerr=error, ecolor='r', alpha=0.5, edgecolor='b', linewidth=2)
```

```python
# Classic Pie Chart /Классическая круговая диаграмма

vals = [24, 17, 53, 21, 35]
labels = ['Ford', 'Toyota', 'BMW', 'AUDI', 'Jaguar']
fig, ax = plt.subplots()
ax.pie(vals, labels=labels)
ax.axis('equal')
```

```python
# An example demonstrating how to work with the parameters of the pie() function
# Пример, демонстрирующий работу с параметрами функции pie()

vals = [24, 17, 53, 21, 35]
labels = ['Ford', 'Toyota', 'BMW', 'AUDI', 'Jaguar']
explode = (0.1, 0, 0.15, 0, 0)
fig, ax = plt.subplots()
ax.pie(vals, labels=labels, autopct='%1.1f%%', shadow=True, 
explode=explode, wedgeprops={'lw':1, 'ls':'--','edgecolor':'k'}, 
rotatelabels=True)
ax.axis('equal')
```

```python
# Pie chart with hole /Круговая диаграмма с отверстием (кольцевая диаграмма)
vals = [24, 17, 53, 21, 35]
labels = ['Ford', 'Toyota', 'BMV', 'AUDI', 'Jaguar']
fig, ax = plt.subplots()
ax.pie(vals, labels=labels, wedgeprops=dict(width=0.5))
```

```python
# 3D plotting. / Построение 3D-графиков.

import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
# line graph / Линейный график
# To plot a line graph, use the plot() function from Axes3D
# Для построения линейного графика используется функция plot() из Axes3D

x = np.linspace(-np.pi, np.pi, 50)
y = x
z = np.cos(x)
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot(x, y, z, label='parametric curve')
```

```python
# Scatter Plot. / Точечный график (Диаграмма рассеяния)
# To build a scatterplot, use the scatter() function from Axes3D 
# Для построения диаграммы рассеяния используется функция scatter() из Axes3D

np.random.seed(123)
x = np.random.randint(-5, 5, 40)
y = np.random.randint(0, 10, 40)
z = np.random.randint(-5, 5, 40)
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x, y, z)
```

```python
# Frame surface / Каркасная поверхность
# To plot a wireframe, use the plot_wireframe() function from Axes3D 
# Для построения каркасной поверхности используется функция plot_wireframe() из Axes3D

u, v = np.mgrid[0:2*np.pi:20j, 0:np.pi:10j]
x = np.cos(u)*np.sin(v)
y = np.sin(u)*np.sin(v)
z = np.cos(v)
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_wireframe(x, y, z)
ax.legend()
```

```python
# Surface / Поверхность
# Для построения поверхности используется функция plot_surface() из Axes3D

u, v = np.mgrid[0:2*np.pi:20j, 0:np.pi:10j]
x = np.cos(u)*np.sin(v)
y = np.sin(u)*np.sin(v)
z = np.cos(v)
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, z, cmap='inferno')
ax.legend()
```
